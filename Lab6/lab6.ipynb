```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, laplace

def plot_dist(dist):
    x = np.linspace(-6, 6, 200)
    # three different parameter settings of a model (red, green, blue)
    pdf1 = dist.pdf(x, 0, 1)
    pdf2 = dist.pdf(x, 1, 3)
    pdf3 = dist.pdf(x, -2.5, 0.5)
    
    fig = plt.figure(figsize=(10,5))
    ax = fig.add_subplot(111)
    
    ax.plot(x, pdf1, color='r', alpha=0.5)
    ax.fill_between(x, pdf1, color='r', alpha=0.3)
    ax.plot(x, pdf2, color='g', alpha=0.5)
    ax.fill_between(x, pdf2, color='g', alpha=0.3)
    ax.plot(x, pdf3, color='b', alpha=0.5)
    ax.fill_between(x, pdf3, color='b', alpha=0.3)
    
    # marginalise out parameter according to our belief p(0 = red) = 0.3.. etc
    # this leads to the evidence in black
    
    # we can think of this as how a model & our beliefs place probability
    # mass over the space where we can later observe data
    
    # e.g. if we observe more data Y, we can evaluate this under the evidence
    # and say 'what is the evidence this model is the right one?'
    pdf4 = 0.3 * pdf1 + 0.2 * pdf2 + 0.5 * pdf3
    ax.plot(x, pdf4, color='k', alpha = 0.8)
    ax.fill_between(x, pdf4, color='k', alpha=0.3)
   
plot_dist(norm)
```

```python
plot_dist(laplace)

# so, if we observed new data Y centered around -1 and evaluated the
# data on the two models, we can say that the data is more probable
# under the laplace compared to the gaussian

# this is cool because it allows us to test different hypothesis about model,
# not just parameters of models
```

```python
import itertools as it

def gen_data(N=3):
    D2 = np.array(list(it.product([-1,1], repeat=N*N)))
    D = []
    for i in range(0, len(D2)):
        d = D2[i]
        D.append(d.reshape([N,N]))
        x = []
    for i in range(-(N-2), N-1):
        for j in range (-(N-2), N-1):
            x.append(np.array([float(i), float(j)]))
    return D, x

gen_data()
```
